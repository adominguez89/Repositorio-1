```python
from tkinter import *
from tkinter import messagebox
from math import sqrt

raiz=Tk()
raiz.title("Calculadora")
raiz.resizable(0,0)
# raiz.iconbitmap("/home/ariel/Documentos/GIT/Prueba/tec.ico")

variable=StringVar()
operacion=StringVar()
valor=StringVar()
nuevo=StringVar()
memoria=StringVar()

edit1=Entry(raiz,width=31,textvariable=variable)
edit1.pack(pady=5,padx=5)
edit1.config(bg="white",fg="black",justify="right")

miFrame=Frame(raiz,width=31,height=240)
miFrame.pack(pady=5,padx=5)

# -------------------- Valores -------------------

def siete():
	var1=operacion.get()
	var2=nuevo.get()

	if (var1!="" and var2=="true"):
		var=""
		variable.set(var+"7")
		var2=""
		nuevo.set(var2)
	else:
		var=variable.get()
		variable.set(var+"7")

def ocho():
	var1=operacion.get()
	var2=nuevo.get()

	if (var1!="" and var2=="true"):
		var=""
		variable.set(var+"8")
		var2=""
		nuevo.set(var2)
	else:
		var=variable.get()
		variable.set(var+"8")

def nueve():
	var1=operacion.get()
	var2=nuevo.get()
	
	if (var1!="" and var2=="true"):
		var=""
		variable.set(var+"9")
		var2=""
		nuevo.set(var2)
	else:
		var=variable.get()
		variable.set(var+"9")

def cuatro():
	var1=operacion.get()
	var2=nuevo.get()

	if (var1!="" and var2=="true"):
		var=""
		variable.set(var+"4")
		var2=""
		nuevo.set(var2)
	else:
		var=variable.get()
		variable.set(var+"4")

def cinco():
	var1=operacion.get()
	var2=nuevo.get()

	if (var1!="" and var2=="true"):
		var=""
		variable.set(var+"5")
		var2=""
		nuevo.set(var2)
	else:
		var=variable.get()
		variable.set(var+"5")

def seis():
	var1=operacion.get()
	var2=nuevo.get()

	if (var1!="" and var2=="true"):
		var=""
		variable.set(var+"6")
		var2=""
		nuevo.set(var2)
	else:
		var=variable.get()
		variable.set(var+"6")

def uno():
	var1=operacion.get()
	var2=nuevo.get()

	if (var1!="" and var2=="true"):
		var=""
		variable.set(var+"1")
		var2=""
		nuevo.set(var2)
	else:
		var=variable.get()
		variable.set(var+"1")

def dos():
	var1=operacion.get()
	var2=nuevo.get()

	if (var1!="" and var2=="true"):
		var=""
		variable.set(var+"2")
		var2=""
		nuevo.set(var2)
	else:
		var=variable.get()
		variable.set(var+"2")

def tres():
	var1=operacion.get()
	var2=nuevo.get()

	if (var1!="" and var2=="true"):
		var=""
		variable.set(var+"3")
		var2=""
		nuevo.set(var2)
	else:
		var=variable.get()
		variable.set(var+"3")

def cero():
	var1=operacion.get()
	var2=nuevo.get()

	if (var1!="" and var2=="true"):
		var=""
		variable.set(var+"0")
		var2=""
		nuevo.set(var2)
	else:
		var=variable.get()
		if var!="0":
			variable.set(var+"0")

def punto():
	var2=nuevo.get()

	if var2!="true":
		var=variable.get()
		if (var.count(".")==0 and len(var)>=1):
			variable.set(var+".")

#  -------------- Operaciones -----------------

def limpiar():
	variable.set("")

def borrar():
	var=variable.get()
	if len(var)>=2:
		var1=var[0:len(var)-1]
		variable.set(var1)
	else:
		variable.set("")

def signo():
	var=variable.get()
	try:
		num=float(var)
		num*=-1
		var=str(num)
		variable.set(var)
	except ValueError:
		messagebox.showinfo(message="No ha introducido un valor.", title="Operación no válida")

def rcuadrada():
	var=variable.get()
	try:
		num=float(var)
		num=sqrt(num)
		var=str(num)
		variable.set(var)
	except ValueError:
		messagebox.showinfo(message="No ha introducido un valor.", title="Operación no válida")

def mplus():
	var=variable.get()
	if var!="":
		memoria.set(var)
		bmplus.config(bg="red")
		messagebox.showinfo(message="Se ha guardado un valor.", title="Memoria activada")
	else:
		messagebox.showinfo(message="No ha introducido un valor.", title="Operación no válida")

def mmenus():
	memoria.set("")
	bmplus.config(bg="gray")
	messagebox.showinfo(message="Se ha borrado la memoria.", title="Memoria desactivada")

def rmemoria():
	var=memoria.get()
	if var!="":
		variable.set(var)
	else:
		messagebox.showinfo(message="La memoria está vacía.", title="Operación no válida")


def inversa():
	var=variable.get()
	try:
		num=float(var)
		num=1/num
		var=str(num)
		variable.set(var)
	except ZeroDivisionError:
		messagebox.showerror(message="No se puede dividir por cero.", title="Operación no definida")
	except ValueError:
		messagebox.showinfo(message="No ha introducido un valor.", title="Operación no válida")

def cuadrado():
	var=variable.get()
	try:
		num=float(var)
		num*=num
		var=str(num)
		variable.set(var)
	except ValueError:
		messagebox.showinfo(message="No ha introducido un valor.", title="Operación no válida")

def suma():
	var1=variable.get()
	valor.set(var1)
	operacion.set("suma")
	nuevo.set("true")

def resta():
	var1=variable.get()
	valor.set(var1)
	operacion.set("resta")
	nuevo.set("true")

def multiplicar():
	var1=variable.get()
	valor.set(var1)
	operacion.set("multiplicar")
	nuevo.set("true")

def dividir():
	var1=variable.get()
	valor.set(var1)
	operacion.set("dividir")
	nuevo.set("true")

def porciento():
	var=variable.get()
	var1=operacion.get()
	var2=valor.get()
	try:
		if var1=="multiplicar":
			num1=float(var2)
			num2=float(var)
			num=num1*num2/100
			var=str(num)
			variable.set(var)
			operacion.set("")
			valor.set("")
	except ValueError:
		messagebox.showinfo(message="No ha introducido un valor.", title="Operación no válida")

def resultado():
	var=variable.get()
	var1=operacion.get()
	var2=valor.get()
	try:
		if var1=="suma":
			num1=float(var2)
			num2=float(var)
			num=num1+num2
			var=str(num)
			variable.set(var)
			operacion.set("")
			valor.set("")
		elif var1=="resta":
			num1=float(var2)
			num2=float(var)
			num=num1-num2
			var=str(num)
			variable.set(var)
			operacion.set("")
			valor.set("")
		elif var1=="multiplicar":
			num1=float(var2)
			num2=float(var)
			num=num1*num2
			var=str(num)
			variable.set(var)
			operacion.set("")
			valor.set("")
		elif var1=="dividir":
			num1=float(var2)
			num2=float(var)
			num=num1/num2
			var=str(num)
			variable.set(var)
			operacion.set("")
			valor.set("")
	except ValueError:
		messagebox.showinfo(message="No ha introducido un valor.", title="Operación no válida")
	except ZeroDivisionError:
		messagebox.showerror(message="No se puede dividir por cero.", title="Operación no definida")


# ------------------------- Fila Cero -------------------------

bmplus=Button(miFrame,text="M+",width=12,height=1,command=mplus)
bmplus.grid(row=0,column=0,padx=1,pady=1,columnspan=2)
bmplus.config(bg="gray")

bmmenus=Button(miFrame,text="M-",width=12,height=1,command=mmenus)
bmmenus.grid(row=0,column=2,padx=1,pady=1,columnspan=2)



# ------------------------- Primera Fila ----------------------

bporciento=Button(miFrame,text="%",width=5,height=2,command=porciento)
bporciento.grid(row=1,column=0,padx=1,pady=1)

blimpiar=Button(miFrame,text="CE",width=5,height=2,command=limpiar)
blimpiar.grid(row=1,column=1,padx=1,pady=1)

bmemoria=Button(miFrame,text="MR",width=5,height=2,command=rmemoria)
bmemoria.grid(row=1,column=2,padx=1,pady=1)

bborrar=Button(miFrame,text="<-",width=5,height=2,command=borrar)
bborrar.grid(row=1,column=3,padx=1,pady=1)

# ------------------------- Segunda Fila ----------------------

binversa=Button(miFrame,text="1/x",width=5,height=2,command=inversa)
binversa.grid(row=2,column=0,padx=1,pady=1)

bcuadrado=Button(miFrame,text="x^2",width=5,height=2,command=cuadrado)
bcuadrado.grid(row=2,column=1,padx=1,pady=1)

braiz=Button(miFrame,text="Raiz",width=5,height=2,command=rcuadrada)
braiz.grid(row=2,column=2,padx=1,pady=1)

bdividir=Button(miFrame,text="/",width=5,height=2,command=dividir)
bdividir.grid(row=2,column=3,padx=1,pady=1)

# ------------------------ Tercera Fila ------------------------

bsiete=Button(miFrame,text="7",width=5,height=2,command=siete)
bsiete.grid(row=3,column=0,padx=1,pady=1)

bocho=Button(miFrame,text="8",width=5,height=2,command=ocho)
bocho.grid(row=3,column=1,padx=1,pady=1)

bnueve=Button(miFrame,text="9",width=5,height=2,command=nueve)
bnueve.grid(row=3,column=2,padx=1,pady=1)

bmultiplicar=Button(miFrame,text="X",width=5,height=2,command=multiplicar)
bmultiplicar.grid(row=3,column=3,padx=1,pady=1)

# ------------------------- Cuarta Fila -----------------

bcuatro=Button(miFrame,text="4",width=5,height=2,command=cuatro)
bcuatro.grid(row=4,column=0,padx=1,pady=1)

bcinco=Button(miFrame,text="5",width=5,height=2,command=cinco)
bcinco.grid(row=4,column=1,padx=1,pady=1)

bseis=Button(miFrame,text="6",width=5,height=2,command=seis)
bseis.grid(row=4,column=2,padx=1,pady=1)

bresta=Button(miFrame,text="-",width=5,height=2,command=resta)
bresta.grid(row=4,column=3,padx=1,pady=1)

# ------------------------ Quinta Fila ------------------

buno=Button(miFrame,text="1",width=5,height=2,command=uno)
buno.grid(row=5,column=0,padx=1,pady=1)

bdos=Button(miFrame,text="2",width=5,height=2,command=dos)
bdos.grid(row=5,column=1,padx=1,pady=1)

btres=Button(miFrame,text="3",width=5,height=2,command=tres)
btres.grid(row=5,column=2,padx=1,pady=1)

bsuma=Button(miFrame,text="+",width=5,height=2,command=suma)
bsuma.grid(row=5,column=3,padx=1,pady=1)

# ------------------ Sexta Fila -----------------------

bsigno=Button(miFrame,text="+/-",width=5,height=2,command=signo)
bsigno.grid(row=6,column=0,padx=1,pady=1)

bcero=Button(miFrame,text="0",width=5,height=2,command=cero)
bcero.grid(row=6,column=1,padx=1,pady=1)

bpunto=Button(miFrame,text=".",width=5,height=2,command=punto)
bpunto.grid(row=6,column=2,padx=1,pady=1)

bigual=Button(miFrame,text="=",width=5,height=2,command=resultado)
bigual.grid(row=6,column=3,padx=1,pady=1)

raiz.mainloop()
```
